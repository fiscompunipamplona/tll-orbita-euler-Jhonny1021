```python
from numpy import array,cos,sin
from numpy import loadtxt
import numpy as np
import scipy as sc
import matplotlib.pyplot as plot
from pylab import plot, show
DTS=1.495*10**11
year=1
m1=1
GMsol=(4*pow(3.1416,2))
TO=year/DTS
deltaT=0.02
n=200
R1=1
v1=2*(3.1416)
the1=0

e_n = [] # M* AU^2/yr^2
x_pos = [] # UA
y_pos = [] # UA
r_pos = []
th_pos = []

def acce_i(r):
    return -GMsol / (r*r*r)

def ener(r):
    return -(GMsol*m) / (2*r)

x_n = R1
y_n = 0.
vx_n = 0.
vy_n = v1

r_n = sqrt(x_n*x_n + y_n*y_n)
th = 0.

e_n.append( ener(r_n) )
x_pos.append( x_n )
y_pos.append( x_n )
r_pos.append( r_n )
th_pos.append( th )
for i in range (0,n):
    tmp_x = x_n
    tmp_y = y_n
    tmp_th = th
    
    
    vx_n = vx_n + tmp_x*acce_i(r_n)*deltaT
    vy_n = vy_n + tmp_y*acce_i(r_n)*deltaT
    
    x_n = tmp_x + vx_n*deltaT
    y_n = tmp_y + vy_n*deltaT
    


    r_n = sqrt(x_n*x_n + y_n*y_n)
    th = tmp_th + ( sqrt(vx_n*vx_n + vy_n*vy_n)/r_n ) * deltaT
        
    e_n.append( ener(r_n) )
    x_pos.append( x_n )
    y_pos.append( y_n )
    r_pos.append( r_n )
    th_pos.append( th )
plt.plot(x_pos,y_pos,'o')
```

```python
plt.polar(th_pos,r_pos,'o')
plt.show()


plt.plot(e_n)
plt.show()
```
